```python
# Define two sets of points
points1 = np.array([[1, 2], [3, 4], [5, 6]])
points2 = np.array([[7, 8], [9, 10], [11, 12]])

print(points1)
print()
print(points2)

# Calculate the Euclidean distance between each pair of points using vectorization
# Euclidean distance between two points (x1, y1) and (x2, y2) is sqrt((x2-x1)^2 + (y2-y1)^2)
# Using vectorization, we can avoid using loops
distances = np.sqrt(np.sum((points2 - points1)**2, axis=1)) # axis=0 is across rows, axis=1 is across cols

# Print the distances
print("Euclidean distances between corresponding points:")
print(distances)
print(distances.shape)
```

```python
# Define three matrices
A = np.random.rand(3, 3)
B = np.random.rand(3, 3)
C = np.random.rand(3, 3)

# Initialize an empty result matrix
result = np.zeros((3, 3))

# Nested for loop to calculate element-wise product
for i in range(3):
    for j in range(3):
        result[i, j] = A[i, j] * B[i, j] * C[i, j]

# Print the result
print("Result using nested loop:")
print(result)

# fast
result_vectorized = np.zeros((3,3))
result_vectorized[:,:] += A[:, :] * B[:, :] * C[:, :]

# Print the result
print("\nResult using vectorization:")
print(result_vectorized)
```

```python
from scipy import integrate
```

```python
# Define the function to integrate
def f(x):
    return x**2

# Define the integration limits
a = 0
b = 1

# Perform the integration
result, error = integrate.quad(f, a, b)

# Print the result
print("Result of the integration:", result)
print("Estimated error:", error)
```

```python
# ANOTHER
# Define the function to integrate
def f(x):
    return np.sin(x)

# Define the integration limits
a = 0
b = np.pi

# Generate sample points for the function
x = np.linspace(a, b, 1000)
y = f(x)

# Perform trapezoidal integration
result = integrate.quad(y, x, 0, 1)

# Print the result
print("Result of trapezoidal integration:", result)
```

```python
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Generate some data points
x = np.linspace(-10, 10, 40)
y = np.sin(x)

# Define the interpolation function
interp_func = interp1d(x, y, kind='cubic')

# Generate new x values for interpolation
x_new = np.linspace(-10, 10, 100)

# Perform interpolation
y_new = interp_func(x_new)

# Plot the original data and interpolated curve
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'o', label='Original data')
plt.plot(x_new, y_new, '-', label='Interpolated curve', linewidth=2)
plt.title('Interpolation with SciPy')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Input a value to find the interpolated value
input_value = 3
interpolated_value = interp_func(input_value)
plt.plot(input_value, interpolated_value, 'ro', label=f'Interpolated value at {input_value:.2f}')
plt.legend()

plt.show()

print(f"Interpolated value at {input_value:.2f}: {interpolated_value:.5f}")
```

```python
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt
from scipy.stats import t


# Sample data
data = np.array(np.loadtxt('linear_regression.txt'))

x = data[:,0]
y = data[:,1]

# Perform linear regression
slope, intercept, _, _, _ = linregress(x, y)

squaredResiduals = (y - np.array(slope * x + intercept)) ** 2
squareDiffFromMean = (y - np.mean(y)) ** 2
R_squared = 1 - (sum(squaredResiduals) / sum(squareDiffFromMean))

df = len(x) - 2
SE_b = np.sqrt( sum(np.array( squaredResiduals / df )) / sum(np.array( (x - np.mean(x)) ** 2 )) )
t_stat = slope / SE_b
p_value = 2 * t.sf(t_stat, len(y) - 2)

# Plot the data points
plt.scatter(x, y, label='Data')

# Plot the linear regression line
plt.plot(x, slope * x + intercept, color='red', label= f"LSRL: ={slope:.2f}x+{intercept:.2f}")

# Add labels and legend
plt.title('Linear Regression Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(0,0,'',label = f'R-squared : {R_squared:0.2f}')
plt.plot(0,0,'',label = f't : {t_stat:.3f}')
plt.plot(0,0,'',label = f'P-value slope : {p_value:.4f}')
plt.legend()

# Show plot
plt.grid(True)
plt.show()

```
